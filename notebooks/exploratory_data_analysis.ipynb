## Exploratory Data Analysis

In this notebook, we will explore the customer data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

First, let's load the data and take a look at the first few rows.

```python
data = pd.read_csv('../data/raw/customer_data.csv')
data.head()
```

Now, let's get some basic statistics about the data.

```python
data.describe()
```

Let's visualize the distribution of each feature.

```python
data.hist(bins=30, figsize=(20,15))
plt.show()
```

Now, let's visualize the correlation between the features.

```python
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True)
```

Finally, let's visualize the pair-wise relationships in the dataset.

```python
sns.pairplot(data)
```
