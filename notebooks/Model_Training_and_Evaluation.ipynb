```python
import pandas as pd
from sklearn.metrics import accuracy_score, f1_score
from src.models.trainable_model import TrainableModel
import os
repo_root = get_repo_root()
# Load all the models
models = [TrainableModel().load(f'{repo_root}/models/{filename[:-4]}_model.pkl') for filename in os.listdir(f'{repo_root}/models') if filename.endswith('_model.pkl')]

# Evaluate all the models
leaderboard = pd.DataFrame()
for model in models:
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    leaderboard = leaderboard.append({'Model': type(model).__name__, 'Accuracy': accuracy, 'F1 Score': f1}, ignore_index=True)

# Sort the leaderboard
leaderboard = leaderboard.sort_values(by='F1 Score', ascending=False)
leaderboard
```
